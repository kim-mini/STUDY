### 2-2. DataFrame

#### 방법 1. list로 만들기

```python
company1 = [['삼성', 2000, '스마트폰'],
           ['현대', 1000, '자동차'],
           ['네이버', 500, '포털']]
```

```python
pd.DataFrame(company1)
```

```python
df1 = pd.DataFrame(company1)
```

```python
df1
```

#### 제목컬럼 만들기

```python
df1.columns = ['기업명', '매출액', '업종']
```

```python
df1
```

#### 방법 2. dict로 만들기

```python
#dictionary
company2 = {'기업명':['삼성', '현대', '네이버'], 
            '매출액': [2000, 1000, 500], 
            '업종': ['스마트폰', '자동차', '포털']
           }
```

```python
df2 = pd.DataFrame(company2)
```

```python
df2
```
